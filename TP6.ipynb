```python
#Question 2. — Produire une figure qui contient les visualisations pour chaque variable explicative. On pourra faire intervenir une boule for, ainsi que data.columns qui donne la liste des noms des variables.

for var in data.columns[:len(data.columns)-1] :
    g = sns.FacetGrid(data=data, hue = 'style')
    g.map(sns.distplot, var).add_legend()
    plt.show()  
```

```python
#Question 3. — En s’appuyant sur les figures obtenues, choisir quatre variables qui semblent les plus prometteuses pour la pr´ediction du type de vin. 
#Cr´eer une DataFrame data reduced ne contenant que ces variables s´electionn´ees ainsi que la variable `a pr´edire. 
data_reduced = data[['volatile_acidity','chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide','style']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_reduced.drop(columns='style'), data_reduced['style'])
```

```python
#Question 4. — Observer les ´echantillons obtenus et d´eterminer le ratio de leurs tailles.
print(y_train.size)
print(y_test.size)
print(y_train.size/(y_train.size+y_test.size))
```

```python
#Question 5. — Afficher les erreurs d’apprentissage et de test du pr´edicteur construit pour la regression logistique. On pensera `a utiliser la fonction f.score().
from sklearn.linear_model import LogisticRegression

f = LogisticRegression()
f.fit(X_train, y_train)
```

```python
print("erreur apprentissage:", 1-f.score(X_train, y_train))
print("erreur de test:", 1-f.score(X_test, y_test))
```

```python
#observer le dataset et la répartition de vins rouge/blanc
len(data_reduced[data_reduced['style']=='red'])/len(data_reduced)
#25% de red, 75% de blanc donc modèle non équilibré!!!
```

```python
#Question 6. — Entraˆıner ´egalement un pr´edicteur Perceptron qu’on pourra charger `a l’aide de la command Perceptron
from sklearn.linear_model import Perceptron
perceptron = Perceptron()

perceptron.fit(X_train, y_train)
```

```python
print("erreur apprentissage:", 1-perceptron.score(X_train, y_train))
print("erreur de test:", 1-perceptron.score(X_test, y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn3 = KNeighborsClassifier(n_neighbors=3)
knn5 = KNeighborsClassifier(n_neighbors=3)

knn3.fit(X_train, y_train)
knn5.fit(X_train, y_train)
```

```python
print("erreur apprentissage pour 3 voisins:", 1-knn3.score(X_train, y_train))
print("erreur de test pour 3 voisins:", 1-knn3.score(X_test, y_test))
print("erreur apprentissage pour 5 voisins:", 1-knn5.score(X_train, y_train))
print("erreur de test pour 5 voisins:", 1-knn5.score(X_test, y_test))
```
